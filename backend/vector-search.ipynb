```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter

# Set encoding explicitly to UTF-8 to avoid Windows decode errors
raw_documents = TextLoader("tagged_description.txt", encoding="utf-8").load()

text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

db_books = Chroma.from_documents(
    documents,
    embedding=embedding_model,
    persist_directory="chroma_db"
)

db_books.persist()
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
# Extract raw token, then clean it by removing quotes if any
isbn_raw = docs[0].page_content.split()[0].strip().strip('"').strip("'")

# Ensure isbn13 column is string
books['isbn13'] = books['isbn13'].astype(str)

# Filter by string match
matched_row = books[books["isbn13"] == isbn_raw]

from IPython.display import display
display(matched_row)
```

```python
def retrieve_semantic_recommendations(query: str, top_k: int = 50) -> pd.DataFrame:
    # Search in vector database
    recs = db_books.similarity_search(query, k=50)

    # Extract ISBNs from page_content
    books_list = []
    for doc in recs:
        try:
            isbn = int(doc.page_content.strip('"').split()[0].replace('.0', ''))
            books_list.append(isbn)
        except ValueError:
            continue

    # Filter the books DataFrame based on ISBNs and return top_k results
    return books[books["isbn13"].astype(int).isin(books_list)].head(top_k).reset_index(drop=True)
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```
